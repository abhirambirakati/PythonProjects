```python
file_name= "atm_data.txt"
balanace = 1000 #Balance is a global Variable that stores money
pin = "1234" #pin is a global variable that Stores ATM pin
#LOAD_DATA
def load_data():
    global balance,pin #global keyword allows us to modify outside variables
    try:
        file = open(file_name, "r") #Opening a file in read mode
        lines = file.readlines() #reads all file into a list
        file.close() # closint the file after reading 
        pin=lines[0].strip() #this line contains pin,and strip is used to remove whiteSpaces or \n like this
        balance=int(lines[1].strip()) #this line contains balance
    except: #if file does not exist or errors occurs
        pass #pass means nothing 
        #Program will use default balance and pin

#Check_Balance
def check_Balance():
    print("Your Balance is:",balance)

#Save_Data
def save_data():
    file = open(file_name, "w")
    file.write(pin + "\n") # write a pin into the file and go to next line
    file.write(str(balance)) #taking balance as String
    file.close() #Closing the file after this

#deposit_money
def deposit_money():
    global balance #global allows us changing  original balance
    try:
        amount = int(input("Enter amount you want to deposit:"))
        balance = balance+amount
        save_data()
        print("money deposited successfully.")
    except:
        print("Please enter numbers only.")
    
#withdraw_money
def withdraw_money():
    global balance
    try:
        withdrawal = int(input("Enter amount to withdraw: "))
        if withdrawal > balance:
            print("Insufficient Balance:")
        else:
            balance = balance-withdrawal
            save_data()
            print("Please collect ur cash.")
    except:
        print("Please Enter number only")

#Changing Pin Of ATM
def change_pin():
    global pin 
    #Asks user old pin if old pin allows is correct then only allows user to rewrite the old pin
    old_pin = input("Enter the Old pin: ")
    if old_pin == pin: #checks if old pin balance
        new_pin = input("Enter New Pin:")
        pin = new_pin
        save_data()
        print("pin changes successfully")
    else:
        print("Incorrect Pin")

#Main Function

def main():
    load_data() #Load data when program starts 
    user_pin = input("Enter The Pin:") #Asks User to enter the pin
    if user_pin != pin: #if pin is wrong it will stop the program  
        print("Incorrect PIN")
        return
    while True:
        print(" \n ----------ATM MENU----------")
        print("1. Check balance")
        print("2. Deposit Money")
        print("3. Withdraw money")
        print("4. Change PIN")
        print("5. EXIT")


        choice = input("Enter Your Choice")
        if choice == "1":
            Check_Balance()
        elif choice == "2":
            deposit_money()
        elif choice == "3":
            withdraw_money()
        elif choice == "4":
            change_pin()
        elif choice == "5":
            print("thank u for using ATM")
            break
        else:
            print("Invalid option")
            
main()       
        
            
            


    
    
    
        
    
        
```
