```python
# '''Start player name  - 
# Load High Score From the file
# Ask player name
# show questions one by one
# take answers
# check answers
# increase score
# after quiz end
# compare with high score
# Save high score
# # end'''
```

```python
# '''
# questions -> List,
# answers,
# score ->number,
# high_score -> File,
# player_name -> String '''
```

```python
#Quiz Game Project
file_name = "high_score.txt"
score = 0
high_score = 0
def load_high_score(): #Load High Score
    global high_score
    try:
        file = open(file_name,"r")
        high_score = int(file.read())
        file.close()
    except:
        pass

#Save High Score
def save_high_score():
    file = open(file_name, "w")
    file.write(str(high_score))
    file.close()

#Play Quiz
def play_quiz():
    global score
    questions = ["what is output of print(2+3)?","Which datatype stores true/false?", "which keyword is used to fetch global variable in the function"]
    answers = ["5", "bool", "global"]
    for i in range(len(questions)):
        print("\nQ.", questions[i])
        user_answer= input("Enter ur answer: ").strip().lower()
        if user_answer == answers[i]:
            score = score+1
            print("Correct")
        else:
            print("Wrong")
def main(): #Main Program
    global high_score #global allows updating same high score
    
    load_high_score()
    name = input("Enter your name")
    play_quiz() #Start Quiz
    print("\n Quiz over")
    print(name, "your score is :", score)
    if score>high_score:
        high_score = score
        load_high_score()
    else:
        print("High Score is:", high_score)
main()
```
